```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
file_path = 'demo.csv'
# https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html
data = pd.read_csv(file_path, sep=',', decimal='.', parse_dates=[1])  
print(data)
```

```python
# https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
time_fmt = mdates.DateFormatter('%H:%M:%S')

fig1, ax1 = plt.subplots(figsize=(10, 3))
ax1.plot(data['Timestamp'], data['X'], label='x', alpha=0.66)
ax1.plot(data['Timestamp'], data['Y'], label='y', alpha=0.66)
ax1.xaxis.set_major_formatter(time_fmt)
ax1.legend()
plt.show()
```

```python
fig2, ax2 = plt.subplots(figsize=(3, 3))
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist2d.html
ax2.hist2d(data['X'], data['Y'], bins=10, cmap='viridis_r', cmin=0.00000005, density=True) 
plt.show()
```
